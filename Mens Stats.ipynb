```python
import cPickle
import csv
import os
import re
import urllib

from bs4 import BeautifulSoup
import pandas as pd
import requests
import seaborn as sns

%matplotlib inline

BASE_URL = "http://play.usaultimate.org"
```

```python
if not os.path.isfile("mens_match_urls.pkl"):
    # Get all match reports
    r = requests.get(BASE_URL + "/events/USA-Ultimate-D-I-College-Championships/schedule/Mens/College-x-College/")
    soup = BeautifulSoup(r.text)
    reports = soup.findAll("a", attrs={"href": re.compile("EventGameId")})
    match_urls = [report.attrs["href"] for report in reports]
    cPickle.dump(match_urls, file("mens_match_urls.pkl", 'w'))

match_urls = cPickle.load(file("mens_match_urls.pkl"))
```

```python
class Stats(object):
    def __init__(self, name, goals, assists, ds, turns, team):
        self.name = name
        self.goals = goals
        self.assists = assists
        self.ds = ds
        self.turns = turns
        self.team = team

    def __add__(self, other):
        return Stats(self.name, self.goals+other.goals,
                self.assists+other.assists, self.ds+other.ds,
                self.turns+other.turns, self.team)
```

```python
def teamseed(team):
    team, seed = team.rsplit(" ", 1)
    seed = int(seed.strip("()"))
    return team, seed

with open('mens_gamestats.csv', 'wb') as csvfile:
    c = csv.writer(csvfile, quoting=csv.QUOTE_MINIMAL)
    c.writerow(["game_id", "name", "team", "seed", "number", "goals", "assists", "ds", "turns"])
                      
    for gameid, game in gamestats.iteritems():
        gameid = urllib.unquote(gameid)
        hometeam, homeseed = teamseed(game["home"]["name"])
        homestats = game["home"]["stats"]
        awayteam, awayseed = teamseed(game["away"]["name"])
        awaystats = game["away"]["stats"]
        
        for _, stat in homestats.iteritems():
            number, name = stat.name.split(' ', 1)
            number = number.strip("#")
            name = name.title()
            c.writerow([gameid, name, hometeam, homeseed, number, stat.goals, stat.assists, stat.ds, stat.turns])
        for _, stat in awaystats.iteritems():
            number, name = stat.name.split(' ', 1)
            number = number.strip("#")
            name = name.title()
            c.writerow([gameid, name, awayteam, awayseed, number, stat.goals, stat.assists, stat.ds, stat.turns])
```

```python
g = sns.lmplot("goals", "turns", teamtots, scatter_kws={"marker":""}, size=9)
ax = g.axes[0][0]
for name, stats in teamtots.iterrows():
    ax.annotate(name, xy=(stats["goals"], stats["turns"]), horizontalalignment='center')
ax.set_xlabel("Goals")
ax.set_ylabel("Turnovers")
ax.set_title("Turnovers vs. Goals in the 2014 USAU College Championships")
#ax.annotate("bananas", xy=(60, 60))
g
```
